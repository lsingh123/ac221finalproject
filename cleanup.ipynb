```python
# making income binary and binning continuous stuff (literally just age lol)
income = []
age = []
with open('cleaned1.csv', 'r', encoding='mac_roman') as fi:
    reader = csv.reader(fi)
    next(reader)
        
    for line in reader:
        income.append(int(line[3]))
        age.append(int(line[8]))
```

```python
import numpy as np
income_q = {1:np.quantile(income, 0.25), 2:np.quantile(income, 0.50), 
            3:np.quantile(income, 0.75), 4:np.quantile(income, 1), 5:np.quantile(income, 0.05)}
age_q = {1:np.quantile(age, 0.25), 2:np.quantile(age, 0.50), 
            3:np.quantile(age, 0.75), 4:np.quantile(age, 1)}
```

```python
with open('cleaned1.csv', 'r', encoding='mac_roman') as fi, open('cleaned3.csv', 'w', encoding='mac_roman') as fo:
    reader = csv.reader(fi)
    writer = csv.writer(fo)
    
    # prepare the headers
    headers = next(reader)
    headers.append('race-binary')
    delete = [29, 30, 31, 32, 33]
    headers = headers[0:29] + headers[34:]
    print(headers)
    writer.writerow(headers)
    
    for line in reader:
        
        # make race binary
        line.append('1' if line[13] == ' 1' else '0')
        
        # bin income - 0 for low income and 1 for high
        # we're treating low income as bottom 5%
        if int(line[3]) < income_q[5]:
            line[3] = '0'
        else:
            line[3] = '1'
        
        # bin age by quartile: 0, 1, 2, 3
        if int(line[8]) < age_q[1]:
            line[8] = '0'
        elif int(line[8]) < age_q[2]:
            line[8] = '1'
        elif int(line[8]) < age_q[3]:
            line[8] = '2'
        else:
            line[8] = '3'
        
        # deal with country of origin (bin all international values together)
        # international (100-554) = 1
        if 100 <= int(line[16]) and int(line[16]) <= 554:
            line[16] = '1'
            
        # remove the obvious stuff 
        line = line[0:29] + line[34:]
        
        writer.writerow(line)
```
