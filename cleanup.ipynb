```python
import csv

dic = 'data_dic.csv'
data = 'cps_2016-08.csv'

with open(dic, 'r', encoding='mac_roman', newline='') as fi:
    reader = csv.reader(fi)
    next(reader)
    index = 0
    for line in reader:
        print(index, line)
        index += 1
```

```python
indeces = [0, 6, 8, 11, 17, 30, 31, 37, 43, 45, 47, 48, 51, 52, 54, 59, 62, 65, 70, 71, 72, 78, 79, 198, 200, 201, 202,
         220, 222, 223, 224, 227, 228, 231, 235, 236, 251, 252]
fields = []
with open(dic, 'r', encoding='mac_roman', newline='') as fi:
    reader = csv.reader(fi)
    next(reader)
    index = 0
    for line in reader:
        if index in indeces:
            fields.append(line[1])
        index += 1
print(fields)
```

```python
with open(data, 'r', encoding='mac_roman') as fi, open('cleaned1.csv', 'w', encoding='mac_roman') as fo:
    reader = csv.reader(fi)
    writer = csv.writer(fo)
    
    for line in reader:
        writer.writerow([line[i] for i in range(len(line)) if i in indeces])
```

```python
with open('cleaned1.csv', 'r', encoding='mac_roman') as fi:
    reader = csv.reader(fi)
    missing_counts = {i:0 for i in range(len(indeces))}
    
    for line in reader:
        missing = sum([1 if line[i] == '' else 0 for i in range(len(line))])
        missing_counts[missing] += 1
    print(missing_counts)
```

```python
# making income binary and binning continuous stuff (literally just age lol)
import csv

income = []
age = []
with open('cleaned1.csv', 'r', encoding='mac_roman') as fi:
    reader = csv.reader(fi)
    
    for line in reader:
        income.append(int(line[3]))
        age.append(int(line[8]))
```

```python
import numpy as np
income_q = {1:np.quantile(income, 0.25), 2:np.quantile(income, 0.50), 
            3:np.quantile(income, 0.75), 4:np.quantile(income, 1)}
age_q = {1:np.quantile(age, 0.25), 2:np.quantile(age, 0.50), 
            3:np.quantile(age, 0.75), 4:np.quantile(age, 1)}
```

```python
with open('cleaned1.csv', 'r', encoding='mac_roman') as fi, open('cleaned2.csv', 'w', encoding='mac_roman') as fo:
    reader = csv.reader(fi)
    writer = csv.writer(fo)
    
    # prepare the headers
    headers = next(reader)
    headers.append('race-binary')
    delete = [30, 32, 33]
    for index in delete:
        headers.pop(index)
    writer.writerow(headers)
    
    for line in reader:
                
        # make race binary
        line.append('white' if line[3] == '1' else 'other')
        
        # bin income - 0 for low income and 1 for high
        if int(line[3]) < income_q[2]:
            line[3] = '0'
        else:
            line[3] = '1'
        
        # bin age by quartile: 0, 1, 2, 3
        if int(line[8]) < age_q[1]:
            line[8] = '0'
        elif int(line[8]) < age_q[2]:
            line[8] = '1'
        elif int(line[8]) < age_q[3]:
            line[8] = '2'
        else:
            line[8] = '3'
        
        # deal with country of origin (bin all international values together)
        # international (100-554) = 1
        if 100 <= int(line[16]) and int(line[16]) <= 554:
            line[16] = '1'
            
        # remove the obvious stuff 
        for index in delete:
            line.pop(index)
        
        writer.writerow(line)
```
